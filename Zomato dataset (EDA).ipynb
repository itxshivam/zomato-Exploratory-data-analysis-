```python
df_contry=pd.read_excel('Country-Code.xlsx')
```

```python
df_contry.head()
```

merge both data df and country_code

```python
df_f=df.merge(df_contry,how='left')
```

```python
df_f.columns
```

```python
df_f['Country']
```

```python
df_f[['Currency','Country']].value_counts().reset_index()
```

```python
country_name=df_f['Country'].value_counts().index
country_value=df_f['Country'].value_counts().values
```

```python
plt.pie(country_value[:3],labels=country_name[:3],autopct="%1.1f%%")
```

94.4% of order in india along .

```python
booking=df_f[['Has Table booking','Country']].value_counts().reset_index().rename(columns={0:'number'})
```

```python
f_book=booking[booking['Has Table booking']=='Yes'].value_counts().reset_index()
```

```python
matplotlib.rcParams['figure.figsize'] = (15,5)
sns.barplot(x='Country',y='number',data=f_book)
```

in case of table booking , india break the record over other country.

```python
ratings=df_f.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'num'})
ratings
```

```python
sns.barplot(x='Aggregate rating',y='num',hue='Rating color',data=ratings,palette=["white","red","orange","yellow","green","green"])
```

mostly people rate in between 3 - 3.5
